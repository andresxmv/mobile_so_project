```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
%matplotlib inline 
```

```python
gdp = pd.read_csv("output/gdp_IMF.csv")
gdp.head()
```

```python
mobile_so = pd.read_csv("output/mobile_so.csv")
mobile_so.head()
```

```python
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Bosnia And Herzegovina"],"Bosnia and Herzegovina")
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Hong Kong"],"China, P.R.: Hong Kong")
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Macao"],"China, P.R.: Macao")
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Croatia"], 'Croatia, Rep. of')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Czech Republic"], 'Czech Rep.')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Estonia"], 'Estonia, Rep. of')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Isreal"], 'Israel')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["South Korea"], 'Korea, Rep. of')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Netherlands"], 'Netherlands, The')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Macedonia"], 'North Macedonia, Republic of')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Poland"], 'Poland, Rep. of')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Serbia"], 'Serbia, Rep. of')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Slovakia (Slovak Republic)"], 'Slovak Rep.')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["Slovenia"], 'Slovenia, Rep. of')
mobile_so["Country Name"]  = mobile_so["Country Name"] .replace(["United States Of America"], 'United States')
```

```python
# Chequear diferencias
set(gdp["Country Name"]).difference(mobile_so["Country Name"])
```

```python
gdp_filter = gdp[(gdp["Country Name"] != "China, P.R.: Mainland") & (gdp["Country Name"] != "Euro Area")]
# Chequear diferencias
set(gdp_filter["Country Name"]).difference(mobile_so["Country Name"])
```

```python
df = gdp_filter.merge(mobile_so, on=["Country Name", "Year", "Quarter"], how="left")
df.head()
```

```python
df["Year_Quarter"] = df["Year_Quarter"].str.replace("Q", "-Q") 
df["Year_Quarter"] = df["Year_Quarter"].str.replace(r'(Q\d) (\d+)', r'\2-\1')
df.head()
```

```python
df["date"] = pd.to_datetime(df["Year_Quarter"])
df.set_index("date",inplace=True)
df["date"] = df.index
df.head()
```

```python
fig = px.line(df, x="date", y="GDP", color='Country Name')
fig.show()
```

```python
print("Cantidad de países: ", len(df["Country Name"].unique()))
print("Número de observaciones: ", df.shape[0])
```
