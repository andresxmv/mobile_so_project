```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import glob
import datetime
%matplotlib inline 
```

```python
path = "data/" # use your path
all_files = glob.glob(path + "/*.csv")
all_files
```

```python
all_df = []
for file in all_files:
    df = pd.read_csv(file)
    all_df.append(df)
```

```python
df = pd.concat(all_df).drop(columns = "Unnamed: 0").reset_index(drop=True)
print(df.shape)
print(df.columns)
df.head()
```

```python
df = df[['Date', "country", 'iOS', 'Android', 'Unknown', 'Windows', 'Samsung',
       'Nokia Unknown', 'BlackBerry OS', 'Linux', 'Other',
       'Series 40', 'SymbianOS', 'Sony Ericsson', 'Playstation', 'LG', 'OS X',
       'Maemo 5', 'bada', 'Nintendo', 'Nintendo 3DS', 'MeeGo', 'JAVA',
       'Firefox OS', 'Tizen', 'WAP', 'webOS', 'LiMo', 'Motorola', 'Palm',
       'KaiOS', 'Rex', 'DangerOS', 'Brew', 'Sailfish']]
df.rename(columns = {"country":"Country Name"}, inplace=True)
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
df.columns 
```

```python
df = df.set_index('Date').groupby('Country Name').resample('Q')['iOS', 'Android', 'Unknown', 'Windows',
       'Samsung', 'Nokia Unknown', 'BlackBerry OS', 'Linux', 'Other',
       'Series 40', 'SymbianOS', 'Sony Ericsson', 'Playstation', 'LG', 'OS X',
       'Maemo 5', 'bada', 'Nintendo', 'Nintendo 3DS', 'MeeGo', 'JAVA',
       'Firefox OS', 'Tizen', 'WAP', 'webOS', 'LiMo', 'Motorola', 'Palm',
       'KaiOS', 'Rex', 'DangerOS', 'Brew', 'Sailfish'].mean().reset_index()
df.head()
```

```python
df["Year"] = pd.DatetimeIndex(df["Date"]).year
df["Quarter"] = pd.DatetimeIndex(df["Date"]).quarter
df.head()
```

```python
df.to_csv("output/mobile_so_monthly.csv", index=None)
```
